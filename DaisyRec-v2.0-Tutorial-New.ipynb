## Tutorial for DaisyRec-v2.0
- This is a basic tutorial for how to use DasiyRec-v2.0
- It includes two parts, i.e., hyper-parameter tuning and test
- Let's take __NeuMF__ runing on __ML-1M__ as an example

### Hyper-parameter Tuning
- **Step 1**: Open [Tune](https://daisyrec.netlify.app/) in the GUI Command Generator
- **Step 2**: Fill in 'Basic Settings' and click the 'Submit' Button

- **Step 3**: Algorithm-Specific Settings
    - Tick specific algorithom parameters
    - Set the parameters searching space, you may select either 'int/float' or 'int choice/float choice'
        -  For int/float, simply set \['Min. Value', 'Max. Value', 'step size'\]
        -  For int/float choice, directly set \['Value'\], separated by comma
    - Click 'Submit' Button

- **Step 4**: Paste the generated command into the terminal to start tune parameter


- **Step 5**: After tuning, the results are automatically saved under './tune_res/best_params_CL_neumf_ml-1m_origin_tloo.csv'
- **Step 6**: The best result of the 30 trails and the corresponding parameter settings are available. Note that topk is set as 10, then NDCG@10 will be the optimization metric for the hyper-parameter optimization

### Test Procedure
- **Step 1**: Open [Test](https://daisyrec.netlify.app/) in the GUI Command Generator
- **Step 2**: Fill in 'Basic Settings' (same as Tune) and click 'Submit' Button
- **Step 3**: Fill in 'Algorithm Specific Settings' with the best parameter settings and click 'Submit' Button


- **Step 4**: Paste the generated test command into the terminal to start test


- **Step 5**: After testing, the results are saved under './res/ml-1m/origin/tloo/CL_neumf_with_0.0uniform_kpi_results.csv'
